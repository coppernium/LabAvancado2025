```python
import pandas as pd
import numpy as np
from scipy import constants
from scipy.stats import linregress

def processar_dados_fotoeletrico(caminho_csv):
    """
    Lê os dados de um arquivo CSV e calcula as variáveis necessárias
    para análise do efeito fotoelétrico.

    Parâmetros
    ----------
    caminho_csv : str
        Caminho para o arquivo CSV contendo as colunas:
        ['lambda', 'sigma', 'Vmin']

    Retorna
    -------
    dict
        Dicionário contendo:
        - df : DataFrame original
        - freq : Frequências calculadas (Hz)
        - freq_err : Erro das frequências (Hz)
        - Emin : Energia mínima (J)
        - h : Valores calculados de h (J*s)
        - h_mean : Média de h (J*s)
        - regressao : Dicionário com resultados do ajuste linear
    """
    # ============================
    # 1. Carregar os dados
    # ============================
    df = pd.read_csv(caminho_csv, sep=',')
    print("Colunas disponíveis:", df.columns.tolist())

    # ============================
    # 2. Calcular frequência e erro
    # ============================
    lambda_m = df['lambda'].values * 1e-9  # nm -> m
    sigma_lambda = df['sigma'].values * 1e-9  # nm -> m

    freq = constants.c / lambda_m  # frequência
    freq_err = constants.c * sigma_lambda / (lambda_m ** 2)  # propagação do erro

    # ============================
    # 3. Energia mínima
    # ============================
    Emin = df['Vmin'].values * constants.e  # eV -> J

    # ============================
    # 4. Calcular h para cada ponto
    # ============================
    h = Emin / freq
    h_mean = np.mean(h)

    print("\nValores calculados de h:", h)
    print("Valor médio de h:", h_mean)

    # ============================
    # 5. Ajuste linear (regressão)
    # ============================
    slope, intercept, r_value, p_value, std_err = linregress(freq, Emin)

    regressao = {
        'slope': slope,           # Inclinação (estimativa de h)
        'intercept': intercept,   # Intercepto
        'r2': r_value ** 2,       # Coeficiente de determinação
        'std_err': std_err,       # Erro padrão da inclinação
        'p_value': p_value
    }

    return {
        'df': df,
        'freq': freq,
        'freq_err': freq_err,
        'Emin': Emin,
        'h': h,
        'h_mean': h_mean,
        'regressao': regressao
    }
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from scipy import constants
import pandas as pd
import numpy as np
from scipy import constants
from scipy.stats import linregress

def processar_dados_fotoeletrico(caminho_csv, saida_csv):
    """
    Lê os dados de um arquivo CSV e calcula as variáveis necessárias
    para análise do efeito fotoelétrico. Salva também os valores
    de h ponto a ponto com incertezas em um CSV.
    """
    import pandas as pd
    import numpy as np
    from scipy import constants
    from scipy.stats import linregress

    # ============================
    # 1. Carregar os dados
    # ============================
    df = pd.read_csv(caminho_csv, sep=',')
    print("Colunas disponíveis:", df.columns.tolist())

    # ============================
    # 2. Calcular frequência e erro
    # ============================
    lambda_nm = df['lambda'].values        # nm
    sigma_lambda_nm = df['sigma'].values   # nm

    lambda_m = lambda_nm * 1e-9            # nm -> m
    sigma_lambda = sigma_lambda_nm * 1e-9  # nm -> m

    freq = constants.c / lambda_m
    freq_err = constants.c * sigma_lambda / (lambda_m ** 2)

    # ============================
    # 3. Energia mínima
    # ============================
    Emin = df['Vmin'].values * constants.e  # eV -> J

    # ============================
    # 4. Calcular h ponto a ponto + erro
    # ============================
    h = Emin / freq
    sigma_E = np.zeros_like(Emin)  # caso não tenha erro em Vmin
    h_err = h * np.sqrt((sigma_E / Emin) ** 2 + (freq_err / freq) ** 2)

    h_mean = np.mean(h)
    h_mean_err = np.sqrt(np.sum(h_err**2)) / len(h)

    print("\nValores calculados de h:", h)
    print("Incertezas de h:", h_err)
    print(f"Valor médio de h = {h_mean:.3e} ± {h_mean_err:.3e}")

    # ============================
    # 5. Ajuste linear (regressão)
    # ============================
    slope, intercept, r_value, p_value, std_err = linregress(freq, Emin)
    regressao = {
        'slope': slope,
        'intercept': intercept,
        'r2': r_value ** 2,
        'std_err': std_err,
        'p_value': p_value
    }

    # ============================
    # 6. Salvar CSV com resultados
    # ============================
    df_out = pd.DataFrame({
        "lambda_nm": lambda_nm,
        "lambda_err_nm": sigma_lambda_nm,
        "h": h,
        "h_err": h_err
    })
    df_out.to_csv(saida_csv, index=False)
    print(f"Arquivo salvo: {saida_csv}")

    return {
        'df': df,
        'freq': freq,
        'freq_err': freq_err,
        'Emin': Emin,
        'h': h,
        'h_err': h_err,
        'h_mean': h_mean,
        'h_mean_err': h_mean_err,
        'regressao': regressao
    }


# ============================
# Exemplo de uso
# ============================
caminho_a = 'a-resultados.csv'
caminho_b = 'b-resultados.csv'

# Processa os dois datasets
resultados_a = processar_dados_fotoeletrico("a-resultados.csv", "saida-h-a.csv")
resultados_b = processar_dados_fotoeletrico("b-resultados.csv", "saida-h-b.csv")

# ======== VARIÁVEIS DATASET A ========
freq_a = resultados_a['freq']
freq_err_a = resultados_a['freq_err']
Emin_a = resultados_a['Emin']
h_mean_a = resultados_a['h_mean']
slope_a = resultados_a['regressao']['slope']

# ======== VARIÁVEIS DATASET B ========
freq_b = resultados_b['freq']
freq_err_b = resultados_b['freq_err']
Emin_b = resultados_b['Emin']
h_mean_b = resultados_b['h_mean']
slope_b = resultados_b['regressao']['slope']

import matplotlib.pyplot as plt
import numpy as np
from scipy import constants

# Dados do dataset A como exemplo (pode repetir para B)
lambda_nm = resultados_a['df']['lambda'].values
h = resultados_a['h']
h_err = resultados_a['h_err']


# Valor exato da constante de Planck
h_exato = constants.h

# ===============================
# PLOT
# ===============================
plt.figure(figsize=(10,6), dpi=300)

# Dados experimentais com barras de erro
plt.errorbar(lambda_nm, h, yerr=h_err, fmt='D', capsize=5,
             label='Valores experimentais de h', color='#e41a1c')

# Linha de referência para h exato
plt.axhline(y=h_exato, color='#1f77b4', linestyle='--', linewidth=2,
            label=r'Constante de Planck: h = $6.626 \cdot 10^{-34}$')
# Eixos e legendas
plt.xlabel(r'Comprimento de onda $\lambda$ (nm)', fontsize=14)
plt.ylabel(r'Constante de Planck $h$ (J·s)', fontsize=14)
plt.title('Medição experimental da constante de Planck LED\' Alta potência', fontsize=16)
plt.grid(True, alpha=0.4)
plt.legend()

plt.tight_layout()
plt.show()
```
